```python
import pandas as pd
import seaborn as sns
planets = sns.load_dataset('planets')
```

# Combination of All Styles

```python
cm = sns.light_palette("green", as_cmap=True)
```

```python
s = planets.head(10).style.background_gradient(cmap=cm)
s.highlight_null(null_color = 'yellow').hide_index().set_caption('10 top data of Planets Dataset').set_precision(1).\
bar(subset = ['orbital_period', 'mass', 'distance']).\
set_properties(**{'border-color': 'white'})
```

# Higlight Style Functions

```python
planets.head(10).style.highlight_null(null_color = 'red')
```

# Gradient and Hide Index function

```python
planets.head(10)[['method','year', 'distance']].sort_values(by = 'distance').style.bar(color = 'lightblue', 
                                                                                       subset = 'distance').hide_index()
```

# Custom Function

```python

def color_below_20_red(value):
    if type(value) == type(''):
        return 'color:black'
    else:
        color = 'red' if value <= 20 else 'black'
        return 'color: {}'.format(color)
    
planets.head(10).style.applymap(color_below_20_red)    
```
